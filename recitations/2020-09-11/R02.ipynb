In astronomy, we often talk about the angular size of objects in the sky. If you hold your arm out and extend your index finger, you will make a triangle (the width of your index finger is one side and the lines connecting your eye to either side of your index finger are the other two sides). What is the angular size (in degrees) of your index finger held at arms length?

```python
index_finger_length = 1.0 # in units of index finger
arm_length = 8.5 # index fingers
index_finger_angle = index_finger_length / arm_length # radian
index_finger_angle
```

```python
index_finger_angle_degrees = index_finger_angle * 180 / pi 
index_finger_angle_degrees
```

```python
thetas = linspace(-pi/4, pi/4, 64)
plot(thetas, tan(thetas), label='tan')
plot(thetas, thetas, label='approx')
xlabel(r'$\theta$')
legend(loc='best')
```

```python
pd.set_option("display.max_rows", 101)
pd.DataFrame({
    'theta': thetas,
    'tan(theta)': tan(thetas),
    'tan(theata) approx': thetas,
    'relative error': (thetas - tan(thetas))/tan(thetas)
})
```

Evidently, as long as $\left| \theta \right| \lesssim 0.17$ the error is below 1%.

## HW 1 Q3

Sirius is the brightest star in the night sky mainly because it is so close to us. The parallax of Sirius is 0.379, and its apparent magnitude is -1.46. Is it intrinsically more luminous or less luminous than the Sun?

$$
M = m - \mu,
$$
where 
$$
\mu = 5 \log_{10} \frac{d}{10 \, \mathrm{pc}}
$$

(Recall $M = 4.76$ for the Sun.)

```python
sirius_distance = 1 / 0.379 # parsecs
sirius_distance
```

```python
sirius_mu = 5*log10(sirius_distance / 10)
sirius_mu
```

```python
sirius_M = -1.46 - sirius_mu
sirius_M
```
